Variaveis utiizadas nas analises:

Age (int): A idade do paciente

Gender (int): Sexo biológico do paciente (Masculino/Feminino)

Heart rate (int): O número de batimentos cardíacos por minuto

Systolic blood pressure (int): A pressão nas artérias quando o coração se contrai

Diastolic blood pressure (int): A pressão nas artérias entre as batidas do coração

Blood sugar (float): O nível de glicose no sangue do paciente

CK-MB (float): Uma enzima cardíaca liberada durante danos ao músculo cardíaco

Troponin (float): Um biomarcador proteico altamente específico para lesão do músculo cardíaco

Result (object): O rótulo de resultado indica se o paciente sofreu ou não um ataque cardíaco.

Esta análise foi comparado qual é a maior incidência de ataque cardíaco entre homens e mulheres, chegamos a conclusão que a maior incidência é em homens.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('ataquecardiaco.csv')

media_colesterol = df.groupby('Sex')['Cholesterol'].mean()

media_colesterol.plot(kind='bar')

plt.title('Média de Colesterol por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Colesterol Médio')

plt.show()
```
